First , we import the libraries required.

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Here we are using a dataset of loans from kaggle.

```python
dataset_loans = pd.read_csv('/content/train_u6lujuX_CVtuZ9i (1).csv')
```

Using head() function we get to know about the attributes of the dataset.

```python
dataset_loans.head()
```

We use certain functions to understand the dataset

```python
dataset_loans.describe()
```

```python
dataset_loans.shape
```

As there are NaN values it is necessary to find them and remove.

```python
dataset_loans.isnull()
```

```python
dataset_loans.isnull().sum()
```

```python
dataset_loans = dataset_loans.dropna()
```

```python
# we check whether all null values are dropped
```

```python
dataset_loans.isnull().sum()
```

Since ML Model understands numbers only , we change the values of the attributes which are strings to meaningful intergers

```python
dataset_loans.replace({'Property_Area':{'Rural':0,'Semiurban':1,'Urban':2},
                      'Married':{'No':0,'Yes':1},
                      'Gender':{'Male':1,'Female':0},
                      "Loan_Status":{'N':0,'Y':1},
                      'Education':{'Graduate':1,'Not Graduate':0},
                      'Self_Employed':{'No':0,'Yes':1}},
                      inplace=True)
```

```python
dataset_loans.head()
```

```python
dataset_loans['Dependents']
```

```python
dataset_loans['Dependents'].value_counts()
```

To reduce the variability of Dependents variable,
 we are assigning a certain value to 3+ for ex: 5

```python
dataset_loans = dataset_loans.replace(to_replace='3+', value=5)
```

```python
dataset_loans.head()
```

Now we can segregate the dataset into prediction variable and the data used for prediction  

```python
X = dataset_loans.drop(columns=['Loan_ID','Loan_Status'],axis=1)
Y = dataset_loans['Loan_Status']
```

```python
# lets see X and Y
print(X)
print(Y)
```

lets split this into test and train data using sklearn

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.15,stratify=Y,random_state=10)
print(X.shape, X_train.shape, X_test.shape)
```

Now we Train the Support Vector Machine Model

```python
classifier = svm.SVC(kernel='linear')
classifier.fit(X_train,Y_train)
```

Since we have trained our model lets evaluate it with test data

```python
X_train_prediction = classifier.predict(X_train)
accuracy_of_training_data = accuracy_score(X_train_prediction,Y_train)
print('The Accuracy we are getting on training dataset is :',accuracy_of_training_data )
print('Accuracy in % :',accuracy_of_training_data*100,'%')
```
